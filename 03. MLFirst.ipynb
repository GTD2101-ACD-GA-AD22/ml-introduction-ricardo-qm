# Mi primer Modelo de Machine Learning

En este JNB se revisará el tema de Mi primer Modelo de Machine Learning.

```python
import pandas as pd

melbourne_file_path = 'https://raw.githubusercontent.com/GTD2101-ACD-GA-AD22/ml-introduction-ricardo-qm/main/data/melb_data.csv'
melbourne_data = pd.read_csv(melbourne_file_path) 

melbourne_data.columns
```

```python
melbourne_data
```

Limpieza de datos. Eliminamos los renglones (las casas) que por cualquier razón les llega a faltar algún dato. Eliminación por renglones.

```python
# dropna drops missing values (think of na as "not available")

melbourne_data = melbourne_data.dropna(axis=0)
```

```python
melbourne_data
```

# Seleccionando el dato a predecir

Para este problema queremos predecir el precio de las casas.

```python
y=melbourne_data.Price

y
```

# Seleccionando los Features.

Estas serían las características que vamos a utilizar para determinar los precios.

```python
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X=melbourne_data[melbourne_features]

X
```

```python
X.describe()
```

# Construyendo el Modelo

A partir de los features y conociendo el dato a predecir podemos entonces construir un posible modelo para encontrar un posible patrón de relación entre ambos.

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")

print(melbourne_model.predict(X.head()))
```
